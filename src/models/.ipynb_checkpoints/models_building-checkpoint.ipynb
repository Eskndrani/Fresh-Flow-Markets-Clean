```python
#merge orders and order_items to get customer metrics
customer_metrics = order_items_info.copy()
slice_of_orders = orders_info[orders_info['id'].isin(customer_metrics['order_id'])]
slice_of_orders = slice_of_orders[['id', 'place_id', 'created', 'total_amount']]
slice_of_orders.rename(columns={'id': 'order_id'}, inplace=True)
customer_metrics = customer_metrics.merge(
    slice_of_orders, 
    on='order_id',
    how='left'
)
#merge customer metrics with places_info
slice= places_info[places_info['id'].isin(customer_metrics['place_id'])]
slice = slice[['id', 'waiting_time', 'vip_threshold']]
slice.rename(columns={'id': 'place_id'}, inplace=True)
customer_metrics = customer_metrics.merge(
    slice, 
    on='place_id',
    how='left'
)
#merge customer metrics with menu_items_info to get rating
slice= menu_items_info[menu_items_info['title'].isin(customer_metrics['title'])]
slice = slice[['title', 'rating']]
customer_metrics = customer_metrics.merge(
    slice, 
    on='title',
    how='left'
)
customer_metrics.head()
```

```python
# ============================================================================
# CUSTOMER CHURN & LOYALTY SCORER - Classification Model
# ============================================================================
# Goal: Identify customers likely to stop ordering based on recent experience
# Trigger automatic SMS offers to at-risk customers
# ============================================================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix, 
                             roc_auc_score, roc_curve, precision_recall_curve, auc)
import matplotlib.pyplot as plt
import seaborn as sns

print("="*80)
print("STEP 1: PREPARE DATA - Aggregate by Customer")
print("="*80)

# Merge orders with order items to get detailed metrics per customer

customer_metrics = order_items_info.merge(
    orders_info[['order_id', 'place_id', 'created', 'total_amount']], 
    on='order_id', 
    how='left'
)

# Merge with places to get waiting time
customer_metrics = customer_metrics.merge(
    places_info[['place_id', 'waiting_time', 'vip_threshold']], 
    on='place_id', 
    how='left'
)

# Merge with menu items to get ratings
customer_metrics = customer_metrics.merge(
    menu_items_info[['title', 'rating']],
    left_on='title',
    right_on='title',
    how='left'
)

print(f"Customer metrics shape: {customer_metrics.shape}")
print(f"Columns available: {customer_metrics.columns.tolist()}")

print("\n" + "="*80)
print("STEP 2: CREATE FEATURES - Independent Variables (IVs)")
print("="*80)

# Aggregate features by customer
customer_features = customer_metrics.groupby('user_id').agg({
    'waiting_time': 'mean',           # Average waiting time
    'rating': 'mean',                  # Average product rating
    'votes': 'sum',                    # Total votes/interactions
    'points_redeemed': 'sum',          # Total points redeemed
    'vip_threshold': 'first',          # VIP threshold
    'discount_amount': 'sum',          # Total discounts received
    'quantity': 'sum',                 # Total items purchased
    'order_id': 'count'                # Total number of orders
}).rename(columns={'order_id': 'total_orders'})

# Fill NaN values
customer_features = customer_features.fillna(customer_features.median())

print(f"\nCustomer features shape: {customer_features.shape}")
print(f"\nFeature Statistics:")
print(customer_features.describe())

print("\n" + "="*80)
print("STEP 3: CREATE TARGET VARIABLE - Dependent Variable (DV)")
print("="*80)

# Calculate repeat purchase behavior
# Define churn: customers with low recent purchase frequency
# We'll analyze purchase patterns over time

# Get orders with dates
orders_with_dates = orders_info.copy()
orders_with_dates['order_date'] = pd.to_datetime(orders_with_dates['created'])
orders_with_dates['order_month'] = orders_with_dates['order_date'].dt.to_period('M')

# Calculate purchases in recent months vs earlier months
recent_cutoff = orders_with_dates['order_date'].quantile(0.75)  # Last 25% of timeline
early_cutoff = orders_with_dates['order_date'].quantile(0.25)   # First 25% of timeline

recent_orders = orders_with_dates[orders_with_dates['order_date'] >= recent_cutoff].groupby('user_id').size()
early_orders = orders_with_dates[orders_with_dates['order_date'] <= early_cutoff].groupby('user_id').size()

# Churn indicator: customers who ordered less in recent period
churn_target = pd.DataFrame({
    'early_orders': early_orders,
    'recent_orders': recent_orders
}).fillna(0)

churn_target['is_churned'] = (churn_target['recent_orders'] < churn_target['early_orders']).astype(int)

print(f"\nChurn Distribution:")
print(churn_target['is_churned'].value_counts())
print(f"Churn Rate: {churn_target['is_churned'].mean():.2%}")

print("\n" + "="*80)
print("STEP 4: MERGE FEATURES WITH TARGET")
print("="*80)

# Merge features with target
model_data = customer_features.join(churn_target['is_churned'], how='inner')
model_data = model_data.dropna()

print(f"Model data shape: {model_data.shape}")
print(f"Data types:\n{model_data.dtypes}")

```
